# Model comparition

## Import libraries

```python
import pandas as pd
```

## Read a dataframe with results

```python
result_mae = pd.read_csv("../results/results_mae.csv")
result_rmse = pd.read_csv("../results/results_rmse.csv")
result_r2 = pd.read_csv("../results/results_r2.csv")
```

```python
compare_table = pd.DataFrame(pd.concat([result_mae['model'],\
                            result_mae.sort_values(by="test")['test'],\
                            result_rmse.sort_values(by='test')['test'],\
                            result_r2.sort_values(by='test')['test']], axis=1,\
                            keys=['model', 'MAE', 'RMSE', 'R2']))
compare_table.sort_values(by="MAE").sort_values(by="RMSE").sort_values(by="R2", ascending=False)
```
